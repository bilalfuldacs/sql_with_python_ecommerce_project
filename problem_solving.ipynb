```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import mysql.connector
import numpy as np

conn = mysql.connector.connect(
    host='localhost',
    user='root',
    password='',
    database='ecommerce'
)
cursor = conn.cursor()

df = pd.read_sql('SELECT DISTINCT customer_city FROM customers', conn)
print(df)
print(df.head())
print(df.count())

#count the number of orders placed in 2017
df = pd.read_sql('SELECT COUNT(order_id) FROM orders where year(order_purchase_timestamp)=2017', conn)

print(df)

#find the total sales per category
df = pd.read_sql('''
    SELECT 
        p.product_category,
        ROUND(SUM(oi.price + oi.freight_value), 2) as total_price
    FROM order_items oi
    JOIN products p ON oi.product_id = p.product_id
    GROUP BY p.product_category
    ORDER BY total_price DESC
''', conn)

print(df)

#percentage of orders paid in installments
df = pd.read_sql('''
    SELECT 
        (COUNT(CASE WHEN payment_installments >=1 THEN 1 END) * 100.0 / 
         COUNT(*)) as installment_percentage
    FROM payments
''', conn)

df

#number of customer from each state
df = pd.read_sql('''
    SELECT 
        customer_state,
        COUNT(customer_id) as customer_count
    FROM customers 
    GROUP BY customer_state
    ORDER BY customer_count DESC
''', conn)

print("Number of customers by state:")
print(df)

df = pd.read_sql('''
    SELECT 
        MONTHNAME(order_purchase_timestamp) as month_name,
  
        COUNT(order_id) as number_of_orders
    FROM orders       
    WHERE YEAR(order_purchase_timestamp) = 2018
    GROUP BY month_name
    ORDER BY month_name
''', conn)

print("Detailed monthly analysis for 2018:")
print(df)
plt.bar(df['month_name'], df['number_of_orders'])
plt.show()

#average number of products per order, grouped by customer city
df = pd.read_sql('''
    WITH order_items_count AS (
        SELECT 
            orders.order_id,
            orders.customer_id,
            COUNT(order_items.order_item_id) as number_of_items
    FROM orders
    JOIN order_items ON orders.order_id = order_items.order_id
    GROUP BY orders.customer_id, orders.order_id
    )
    SELECT 
        customer_city,
        AVG(order_items_count.number_of_items) as avg_items_per_order
    FROM customers
    JOIN order_items_count ON order_items_count.customer_id = customers.customer_id
    GROUP BY customer_city
    
''', conn)
df

#percentage of total revenue contributed by each product category
df = pd.read_sql('''
    SELECT 
        p.product_category,
        ROUND(SUM(oi.price + oi.freight_value), 2) as total_price
    FROM order_items oi
    JOIN products p ON oi.product_id = p.product_id
    GROUP BY p.product_category
    ORDER BY total_price DESC
''', conn)
print(df)

#identify the correlation between product price and the number of times a product has been purchased
df = pd.read_sql('''
    SELECT 
        p.product_category,
        AVG(oi.price) as average_price,
        COUNT(oi.order_item_id) as number_of_purchases
    FROM order_items oi
    JOIN products p ON oi.product_id = p.product_id
    GROUP BY p.product_category
    ORDER BY average_price DESC
''', conn)
df

np.corrcoef(df['average_price'], df['number_of_purchases'])

#calculate the total revenue generated by each seller, and rank them by revenue
df = pd.read_sql('''
    SELECT 
        s.seller_id,
        
        ROUND(SUM(oi.price + oi.freight_value), 2) as total_revenue,
                 RANK() OVER (ORDER BY SUM(oi.price + oi.freight_value) DESC) as revenue_rank
    FROM sellers s
    JOIN order_items oi ON s.seller_id = oi.seller_id
    GROUP BY s.seller_id
    ORDER BY total_revenue DESC
''', conn)

print(df)
```
